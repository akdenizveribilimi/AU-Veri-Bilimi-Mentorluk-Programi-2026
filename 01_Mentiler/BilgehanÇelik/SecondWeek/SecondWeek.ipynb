KOŞULLU FİLTRELEME - filter(), map()

```python
recon_results = ["192.168.1.41", "HTTPError:403", "192.168.1.94", 
                "192.168.1.56", "192.168.2.12", "Error", "192.168.2.32"]
```

```python
ips = list(filter(lambda ip: ip.startswith("192.168"), recon_results))
```

```python
ips_status = list(map(lambda ip: (ip, ip.startswith("192.168.1")), ips))
```

```python
lan_ips_status = list(filter(lambda pair: pair[1], ips_status))
```

```python
print("IPs in LAN:")
for ip, ip_status in lan_ips_status:
    print(ip)
```
